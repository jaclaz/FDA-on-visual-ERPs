```python
path_dataframe = os.path.join(home_path, 'cat_dataframe.csv')
cat_dataframe = pd.read_csv(path_dataframe,index_col=0)
```

Nuovo dizionario per gli events con i concepts associati alle macro-categorie

```python
new_dict=cat_dataframe.T.to_dict(orient='list')
clean_dict={}
```

```python
new_dict
```

Questo dizionario ha gli elementi salvati come float e un sacco di nan inutili, quindi lo correggo

```python
#for key_name in new_dict:
    
    #Creo items_cleaned tenendo solo gli elementi non nulli di ogni categoria
#    items_cleaned = [item for item in new_dict[key_name] if not np.isnan(item)]
    
    #Inserisco nel nuovo dizionario clean_dict la categoria come key a cui associo gli elementi appena ottenuti
#    clean_dict[key_name] = items_cleaned
    
    #Trasformo gli elementi ottenuti da float a int
#    clean_dict[key_name] = [int(val) for val in clean_dict[key_name]]

for key_name in new_dict:
    
    #Creo items_cleaned tenendo solo gli elementi non nulli di ogni categoria
    items_cleaned = []
    for item in new_dict[key_name]:
        if not np.isnan(item):
            items_cleaned.append(item)
    
    #Inserisco nel nuovo dizionario clean_dict la categoria come key a cui associo gli elementi appena ottenuti
    clean_dict[key_name] = items_cleaned
    
    temp=[]
    #Trasformo gli elementi ottenuti da float a int
    for val in clean_dict[key_name]:
        temp.append(int(val))
    clean_dict[key_name] = temp
```

Stampo le categorie che ho ottenuto

```python
clean_dict.keys()
```

Associo ad ogni epoch un nuovo numero che mi identificherà la categoria dello stimolo

```python
j=0

#Ciclo for per ogni window aiutandomi con la funzione ausiliaria column() che ho creato
for ii in column(new_events,2):
    k=0
    
    #Ciclo for per ogni categoria
    for key_name in clean_dict:
        
        #Se l'indice dello stimolo (ii) è tra i concepts associati alla categoria che sto considerando allora lo
        #cambio con un numero (k)
        if ii in clean_dict[key_name]:
            new_events[j][2]=k
            
        #Aumento k per non avere categorie con lo stesso indice
        k+=1
    
    #Aumento j per scorrere tutte le windows del paziente
    j+=1 
```

Ora tutte le windows hanno come indice un numero da -1 a 27

```python
new_events
```

Associo ad ogni categoria un numero scorrendo tutte le categorie e aumentando di volta in volta k, così che corrispondando ai nuovi indici

```python
k=0

for key_name in clean_dict:
    clean_dict[key_name]=k
    k+=1

#Il dataset associa agli stimoli di trial l'indice -1, quindi lo inserisco così anche nel nuovo dizionario
clean_dict['trial']=-1
```

Imposto i nuovi id degli eventi
