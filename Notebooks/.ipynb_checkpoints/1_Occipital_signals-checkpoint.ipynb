# Media dei segnali occipitali

```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
home_path = os.path.abspath(os.getcwd())
```

```python
for cat in ['animal', 'food', 'vehicle', 'tool', 'body']:
    occ=pd.DataFrame()
    for idx in np.arange(44):
        data=pd.DataFrame()
            
        for ch in ['O1','Oz','O2','PO7','PO3','POz','PO4','PO8']:
            df=pd.read_csv(home_path+'\\dataframes\\'+cat+'\\'+cat+'_'+ch+'.csv')
                
            if data.empty:
                data['time']=df['time']
                
            if 'sub '+str(idx) in df.columns:
                data[ch] = df['sub '+str(idx)]
                
        if occ.empty:
            occ['time']=data['time']
        occ['sub '+ str(idx)] = data.mean(axis=1)
        
    occ.to_csv('dataframes\\'+cat+'\\'+cat+'_avg.csv')
    
```

```python
cat='animal'
signal = pd.read_csv(home_path+'\\dataframes\\'+cat+'\\'+cat+'_avg.csv')

axes = signal.plot.line(x='time', subplots=True, title=list(signal.columns[1:]), layout = [5,9], figsize=[20,10])
    
```
